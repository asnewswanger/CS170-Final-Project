# Second New Task - T-Test: Comparing if Experience Level significantly impacts Salary in USD

```python
# Pulling all of the values for each category I want to compare
```

```python
#Senior Level Exp Salaries in USD
pd.set_option('display.max_rows', None)
SE_filtered_salaries = data[data['experience_level'] == 'SE']['salary_in_usd']
SE_filtered_salaries
```

```python
# Middle Level Exp Salaries in USD
pd.set_option('display.max_rows', None)
MI_filtered_salaries = data[data['experience_level'] == 'MI']['salary_in_usd']
MI_filtered_salaries
```

```python
t_statistic, p_value = stats.ttest_ind(MI_filtered_salaries, SE_filtered_salaries, equal_var=True) 
# equal_var=True assumes equal variances; set to False for Welch's t-test

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the salaries of workers with Middle Level Experience and Senior Level Experience.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the salaries of workers with Middle Level Experience and Senior Level Experience.")
```
